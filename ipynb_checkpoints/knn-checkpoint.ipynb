```python
%matplotlib inline
import time
import psutil
import os
import matplotlib.pyplot as plt

def memory_usage_psutil():
    # 使用 psutil 获取当前进程的内存占用情况
    process = psutil.Process(os.getpid())
    mem_info = process.memory_info()
    return mem_info.rss / (1024**2), mem_info.vms / (1024**2)  # 返回以MB为单位的物理和虚拟内存占用

def monitor_memory_during_model_computation(rounds=10, interval=1):
    # 存储每一轮物理内存和虚拟内存占用的数据
    physical_memory_usage = []
    virtual_memory_usage = []

    for i in range(rounds):
        # 模型计算过程
        # ... 包括模型加载、数据处理等部分
        # 这里使用sleep来模拟重复任务
        print(i)
        if i < 10:
            time.sleep(interval)
        else:
            # 我们在此处使用time.sleep来模拟一个重复的任务
            mname = f'service_c5_v100_iter20_fast'
            model = load_model(mname)
            # Retrieve embeddings
            dataset = last_day[['ip', 'class']].drop_duplicates()
            embeddings = get_scaled_embeddings(dataset, model, mname, False)
            embeddings.to_csv(f'{DATASETS}/embeddings.csv.gz')
            # Run the kNN classifier
            X, y, X_test, y_test = split_train_test(embeddings, with_unknown=True)
            y_true = y_test
        
        # 记录当前的物理内存和虚拟内存占用
        physical_mem, virtual_mem = memory_usage_psutil()
        physical_memory_usage.append(physical_mem)
        virtual_memory_usage.append(virtual_mem)
        
        time.sleep(interval)

    return physical_memory_usage, virtual_memory_usage


# 开始监测
rounds = 20  # 监测轮数
interval = 5  # 每轮间隔时间（秒）
physical_memory, virtual_memory_usage = monitor_memory_during_model_computation(rounds, interval)

from matplotlib import rcParams

config = {
    "font.family": 'serif', # 衬线字体
    "font.size": 12, # 相当于小四大小
    "font.serif": ['SimSun'], # 宋体
    "mathtext.fontset": 'stix', # matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
    'axes.unicode_minus': False # 处理负号，即-号
}
rcParams.update(config)

# 绘制物理内存占用情况的折线图
plt.figure(1)
plt.plot(range(1, rounds + 1), physical_memory, marker='o')
plt.title('物理内存占用情况')
plt.xlabel('计算次数')
plt.ylabel('占用大小(MB)')
plt.xticks(range(1, rounds + 1))
plt.grid(True)
plt.legend()
plt.show()

# 绘制虚拟内存占用情况的折线图
plt.figure(2)
plt.plot(range(1, rounds + 1), virtual_memory_usage, marker='o')
plt.title('虚拟内存占用情况')
plt.xlabel('计算次数')
plt.ylabel('占用大小(MB)')
plt.xticks(range(1, rounds + 1))
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
%matplotlib inline

def plot_confusion_matrix(confusion_mat):
    # 画混淆矩阵图，配色风格使用cm.Greens
    plt.imshow(confusion_mat,interpolation='nearest',cmap=plt.cm.Greens)
    
    # 显示colorbar
    plt.colorbar()
    
    # 使用annotate在图中显示混淆矩阵的数据
    for x in range(len(confusion_mat)):
        for y in range(len(confusion_mat)):
            plt.annotate(confusion_mat[x, y], xy=(x, y), horizontalalignment='center', verticalalignment='center')
            # 第一个参数是注释的内容
            # xy设置箭头尖的坐标
            # horizontalalignment水平对齐
            # verticalalignment垂直对齐
            # 其余常用参数如下：
            # xytext设置注释内容显示的起始位置
            # arrowprops 用来设置箭头
            # facecolor 设置箭头的颜色
            # headlength 箭头的头的长度
            # headwidth 箭头的宽度
            # width 箭身的宽度

    plt.title('Confusion Matrix')     # 图标title
    plt.ylabel('True label')          # 坐标轴标签
    plt.xlabel('Predicted label')     # 坐标轴标签
    
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, tick_marks)
    plt.yticks(tick_marks, tick_marks)
    plt.show()

C2= confusion_matrix(y_true, y_pred)
print(C2)
plot_confusion_matrix(C2)
```
